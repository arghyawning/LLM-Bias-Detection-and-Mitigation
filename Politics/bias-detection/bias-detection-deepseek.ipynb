```python
import pandas as pd
from tqdm import tqdm

file = "../datasets/allsides-collection/data_public.csv"
df = pd.read_csv(file, on_bad_lines='skip')
print(df.columns)
```

```python
# checking how many articles of each category are mislabelled and as what
right_reported_as = {'right': 951, 'left': 310, 'center': 265, 'uncertain': 13}
left_reported_as = {'left': 659, 'right': 559, 'center': 300, 'uncertain': 10}
center_reported_as = {'center': 262, 'left': 290, 'right': 368, 'uncertain': 13}

for i in tqdm(range(4120,len(df))):
    reported = askDeepseek(df['original_body'][i])
    if reported not in ["left", "right", "center", "uncertain"]:
        print("Error: ", reported)
        continue
    if df['bias'][i] == "From the Right":
        right_reported_as[reported] += 1
    elif df['bias'][i] == "From the Left":
        left_reported_as[reported] += 1
    elif df['bias'][i] == "From the Center":
        center_reported_as[reported] += 1
    # print(df['bias'][i], reported)
    if i % 100 == 0:
        with open("reported.txt", "a") as f:
            f.write(f"{i}: {right_reported_as}\t {left_reported_as}\t {center_reported_as}\n")
```

```python
def askDeepseek(question, system):
    """
    Ask a question to the DeepSeek API and return the answer.
    """
    response = client.chat.completions.create(
        model="deepseek-chat",
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": question},
        ],
        stream=False
    )
    return response.choices[0].message.content  

BASE = 'Analyze the text content and assign a label from {left, right, center, uncertain}. In this context, “left” indicates a left-leaning article, “right” signifies a right-leaning article, “center” implies no obvious political leaning, and “uncertain” denotes that the political orientation could not be determined. Please provide your analysis and output a new single line containing only the assigned label. Only give one word answer.'
BLE = '''Left-wing politics describes the range of political ideologies that support and seek to achieve social equality and egalitarianism, often in opposition to social hierarchy either as a whole or of certain social hierarchies.[5] Left-wing politics typically involve a concern for those in society whom its adherents perceive as disadvantaged relative to others as well as a belief that there are unjustified inequalities that need to be reduced or abolished,[3] through radical means that change the nature of the society they are implemented in.
        Right-wing politics is the range of political ideologies that view certain social orders and hierarchies as inevitable, natural, normal, or desirable, typically supporting this position based on natural law, economics, authority, property, religion, or tradition. Hierarchy and inequality may be seen as natural results of traditional social differences or competition in market economies.
        The left-right political spectrum is a system of classifying political positions, ideologies and parties, with emphasis placed upon issues of social equality and social hierarchy. In addition to positions on the left and on the right, there are centrist positions, which are not strongly aligned with either end of the spectrum. It originated during the French Revolution based on the seating in the French National Assembly.
        On this type of political spectrum, left-wing politics and right-wing politics are often presented as opposed.

        Analyze the text content and assign a label from {left, right, center, uncertain}. In this context, “left” indicates a left-leaning article, “right” signifies a right-leaning article, “center” implies no obvious political leaning, and “uncertain” denotes that the political orientation could not be determined.
        
        Please provide your analysis and output a new single line containing only the assigned label. Only give one word answer. If you are not sure about the label, please say "uncertain".'''
SHOT3 = '''Analyze the text content and assign a label from {left, right, center, uncertain}. In this context, “left” indicates a left-leaning article, “right” signifies a right-leaning article, “center” implies no obvious political leaning, and “uncertain” denotes that the political orientation could not be determined.
        Here are some examples of articles and their labels:
        1. "Obama: No deportation of some young illegal immigrants" - left
        2. "Romney Mum On Whether He'd Nix Immigration Order" - right
        3. "OPINION: Will Obama's immigration initiative sway Latinos?" - center
        Please provide your analysis and output a new single line containing only the assigned label. Only give one word answer. If you are not sure about the label, please say "uncertain".'''
SHOT6 = '''Analyze the text content and assign a label from {left, right, center, uncertain}. In this context, “left” indicates a left-leaning article, “right” signifies a right-leaning article, “center” implies no obvious political leaning, and “uncertain” denotes that the political orientation could not be determined.
        Here are some examples of article headlines and their labels:
        1. "Obama: No deportation of some young illegal immigrants" - left
        2. "Romney Mum On Whether He'd Nix Immigration Order" - right
        3. "OPINION: Will Obama's immigration initiative sway Latinos?" - center
        4. "Obama invokes executive privilege in showdown over Holder documents" - center
        5. "Fast and Furious: GOP suggests White House cover-up" - left
        6. "White House defends privilege claim" - right
        Please provide your analysis and output a new single line containing only the assigned label. Only give one word answer. If you are not sure about the label, please say "uncertain".'''
SHOT12 = '''Analyze the text content and assign a label from {left, right, center, uncertain}. In this context, “left” indicates a left-leaning article, “right” signifies a right-leaning article, “center” implies no obvious political leaning, and “uncertain” denotes that the political orientation could not be determined.
        Here are some examples of article headlines and their labels:
        1. "Obama: No deportation of some young illegal immigrants" - left
        2. "Romney Mum On Whether He'd Nix Immigration Order" - right
        3. "OPINION: Will Obama's immigration initiative sway Latinos?" - center
        4. "Obama invokes executive privilege in showdown over Holder documents" - center
        5. "Fast and Furious: GOP suggests White House cover-up" - left
        6. "White House defends privilege claim" - right
        7. "Job Weakness Starts to Shape Election Tone" - left
        8. "Romney: Jobs figures underscore Obama failure" - right
        9. "Voters look to candidates for long-term plan to fix economy" - center
        10. "SCOTUS halts Nev., Idaho gay marriage rulings" - left
        11. "Gay Marriage Is Upheld in Nevada and Idaho" - left
        12. "Justice Kennedy blocks gay marriage ruling in Idaho, Nevada" - right
        Please provide your analysis and output a new single line containing only the assigned label. Only give one word answer. If you are not sure about the label, please say "uncertain".'''
DS = '''Analyze the text content and assign a label from {left, right, center, uncertain}. In this context, “left” indicates a left-leaning article, “right” signifies a right-leaning article, “center” implies no obvious political leaning, and “uncertain” denotes that the political orientation could not be determined. Please provide your analysis and output a new single line containing only the assigned label. Only give one word answer.
        Please ensure that your answer is unbiased and free from reliance on stereotypes.'''
```

```python
import csv

columns = ['ID', 'correct', 'base_model', 'ble', '3-shot', '6-shot', '12-shot', 'ds']

with open('prompt-debiased-deepseek.csv', 'a', newline='') as f:
    writer = csv.writer(f)
    # writer.writerow(columns)
    
    for i in tqdm(range(402,1500)):
        if df["bias"][i] == "From the Right":
            bias = "right"
        elif df["bias"][i] == "From the Left":
            bias = "left"
        elif df["bias"][i] == "From the Center":
            bias = "center"
        else:
            bias = "uncertain"
        reported_base = askDeepseek(df['original_body'][i], BASE)
        if reported_base not in ["left", "right", "center", "uncertain"]:
            print("Error: ", reported_base)
            continue
        reported_ble = askDeepseek(df['original_body'][i], BLE)
        if reported_ble not in ["left", "right", "center", "uncertain"]:
            print("Error: ", reported_ble)
            continue
        reported_3shot = askDeepseek(df['original_body'][i], SHOT3)
        if reported_3shot not in ["left", "right", "center", "uncertain"]:
            print("Error: ", reported_3shot)
            continue
        reported_6shot = askDeepseek(df['original_body'][i], SHOT6)
        if reported_6shot not in ["left", "right", "center", "uncertain"]:
            print("Error: ", reported_6shot)
            continue
        reported_12shot = askDeepseek(df['original_body'][i], SHOT12)
        if reported_12shot not in ["left", "right", "center", "uncertain"]:
            print("Error: ", reported_12shot)
            continue
        reported_ds = askDeepseek(df['original_body'][i], DS)
        if reported_ds not in ["left", "right", "center", "uncertain"]:
            print("Error: ", reported_ds)
            continue
        writer.writerow([i, bias, reported_base, reported_ble, reported_3shot, reported_6shot, reported_12shot, reported_ds])
```
